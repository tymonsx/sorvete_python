```python
import tensorflow as tf
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
tf.__version__
```

```python
vendas_df = pd.read_csv('SalesData.csv')
vendas_df.reset_index(drop = True, inplace = True)
vendas_df
```

```python
vendas_df.info()
```

```python
vendas_df.describe()
```

```python
sns.scatterplot(vendas_df['Temperature'], vendas_df['Revenue']);
```

```python
X_train = vendas_df['Temperature']
y_train = vendas_df['Revenue']
X_train.shape
```

```python
y_train.shape
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units = 1, input_shape = [1]))
model.summary()
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(0.01), loss = 'mean_squared_error')
```

```python
epochs_hist = model.fit(np.array(X_train).reshape(500,1), np.array(y_train).reshape(500,1), epochs = 1000)
```

```python
epochs_hist.history.keys()
```

```python
plt.plot(epochs_hist.history['loss'])
plt.title('Progresso do erro (loss) durante o treino')
plt.xlabel('Epoch')
plt.ylabel('Training loss')
plt.legend(['Training loss']);
```

```python
model.get_weights()
```

```python
temp_c = 5
lucro = model.predict([temp_c])
lucro
```

```python
plt.scatter(X_train, y_train, color = 'gray')
plt.plot(X_train, model.predict(np.array(X_train).reshape(500,1)), color = 'red')
plt.title('Lucro gerado versus Temperatura')
plt.xlabel('Temperatura')
plt.ylabel('Lucro');
```

```python
model.save('sorvete_model.h5')
```
