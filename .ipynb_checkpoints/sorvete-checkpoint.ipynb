# Etapa 4: Criação das variáveis da base de dados

```python
# Criação das variáveis X_train e y_train
X_train = vendas_df['Temperature']
y_train = vendas_df['Revenue']
```

```python
# Formato da variável X_train
X_train.shape
```

```python
# Formato da variável y_train
y_train.shape
```

# Etapa 5: Criação e construção do modelo

```python
# Construção do modelo sequencial
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units = 1, input_shape = [1]))
```

```python
# Sumário do modelo
model.summary()
```

```python
# Compilar o modelo
model.compile(optimizer=tf.keras.optimizers.Adam(0.01), loss = 'mean_squared_error')
```

```python
# Treinamento
epochs_hist = model.fit(np.array(X_train).reshape(500,1), np.array(y_train).reshape(500,1), epochs = 1000)
```

# Etapa 6: Avaliação do modelo 

```python
# Visualização do dicionário com os resultados
epochs_hist.history.keys()
```

```python
# Gráfico com os resultados da loss function
plt.plot(epochs_hist.history['loss'])
plt.title('Model Loss Progress During Training')
plt.xlabel('Epoch')
plt.ylabel('Training loss')
plt.legend(['Training loss']);
```

```python
# Visualização dos pesos
model.get_weights()
```

```python
# Previsões com o modelo treinado, com a temperatura de 5 graus
temp_c = 5
lucro = model.predict([temp_c])
lucro
```

```python
# Gráfico com a reta da regressão linear
plt.scatter(X_train, y_train, color = 'gray')
plt.plot(X_train, model.predict(np.array(X_train).reshape(500,1)), color = 'red')
plt.title('Lucro gerado versus Temperatura')
plt.xlabel('Temperatura')
plt.ylabel('Lucro');
```
